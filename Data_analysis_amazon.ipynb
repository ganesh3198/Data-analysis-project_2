# DATA  VISUALIZATIONS

# category

```python
import seaborn as sns
```

```python
ax=sns.countplot(x='Category',data=df)
sns.set(rc={'figure.figsize':(40,40)})
for bars in ax.containers:
    ax.bar_label(bars)
```

# group by category  vs quantity

```python
ptr=df.groupby(['Category'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)

ad=sns.barplot(x='Category',y='Qty',data=ptr)
for bars in ad.containers:
    ad.bar_label(bars)
```

# group by category vs amount

```python
ptr=df.groupby(['Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

ad=sns.barplot(x='Category',y='Amount',data=ptr)
for bars in ad.containers:
    ad.bar_label(bars)
```

# size

```python
ax=sns.countplot(x='Size',data=df)
sns.set(rc={'figure.figsize':(60,60)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x='ship-state',data=df)
sns.set(rc={'figure.figsize':(60,200)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ptr=df.groupby(['ship-state'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

ad=sns.barplot(x='ship-state',y='Amount',data=ptr)
for bars in ad.containers:
    ad.bar_label(bars)
```

# groupby size vs quantity

```python
ctr=df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty', data=ctr)
```

# groupby size vs Amount

```python
ctr=df.groupby(['Size'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ds=sns.barplot(x='Size',y='Amount', data=ctr)

sns.set(rc={'figure.figsize':(20,10)})
for bars in ds.containers:
    ds.bar_label(bars)
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10, 5))
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

# B2B using pie chart

```python
B2B_Check = df['B2B'].value_counts()
plt.pie(B2B_Check, labels=B2B_Check, autopct='%.20f%%')
plt.figure(figsize=(20,5))
#plt.axis('equal')
plt.show()
```

```python
B2B_Check = df['B2B'].value_counts()
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
#  Prepare data for pie chart
a1 = df['Fulfilment'].value_counts()

# Step 4: Plot the pie chart
fig, ax = plt.subplots()

ax.pie(a1, labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect="equal")

plt.show
```

# ship_state

```python
plt.figure(figsize=(12, 6))
ax=sns.countplot(data=df, x='ship-state')
for i in ax.containers:
    ax.bar_label(i)
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
x_data = df['Status']  
y_data = df['ship-state'] 
# Plot the scatter plot
plt.scatter(x_data, y_data)
plt.figure(figsize=(50, 5))
plt.xlabel('status ')  
plt.ylabel('ship_city')  
plt.title('Scatter Plot') 
plt.show()
```

# status

```python
df['Status'] = df['Status'].astype(str)
data = df['Status']
plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, edgecolor='red',color='orange')
plt.xticks(rotation=90)
plt.title('status of delivery\n\n',
          fontweight = "bold")

 
plt.show()
```

```python
top_10_state = df['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sn=df[df['ship-state'].isin(top_10_state.index)]
sns.countplot(data=sn, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```
